A cursor is a type of temporary memory or workstation. It's allocated by the database server when a user performs DML (Data Manipulation Language) actions on a table. A SQL cursor is a database object that is used to retrieve data from a result set one row at a time.

### <b>1. Implicit Cursors:</b>
Implicit Cursors are also known as Default Cursors of SQL SERVER. These Cursors are allocated by SQL SERVER when the user performs DML operations.
### <b>2.Explicit Cursors :</b>
Explicit Cursors are Created by Users whenever the user requires them. Explicit Cursors are used for Fetching data from Table in Row-By-Row Manner.

### <b>1. execute():</b>

This method executes an SQL statement.

<b>Syntax:</b>

connection_name=sqlite3.connect('database.db')<br>
connection_name.execute("sql query")

* ### Follow the given steps to use cursor to create a table restaurant and insert records into it.

1. import sqlite3 and connect to a database
2. execute create command by using above mentioned column names.
3. execute insert command to insert various records.
4. execute select command to retrieve the data.
5. apply a for loop to display data row wise.

Refer: https://www.mcobject.com/docs/Content/Programming/Python/Classes/Cursor/execute.htm

```python
# importing sqlite3 module
import sqlite3

# create connection by using object
# to connect with cursors_1 database
connection = sqlite3.connect('cursors_1.db')

# query to create a table named restaurant
connection.execute(''' create table restaurant(Order_ID INT primary key not null, 
                                               Date DATE not null, 
                                               Items varchar(255) not null, 
                                               Bill_amount INT not null, 
                                               Parcel_or_Dine_in varchar(255) not null)''')

# insert query to insert food details in the above table
connection.execute(''' insert into restaurant values(1, '2022-04-01', 'pizza,fries,juice', 550, 'dine-in'),
                                                    (2, '2022-04-01', 'pizza,cake,sandwich,juice', 750.5, 'dine-in'),
                                                    (3, '2022-04-01', 'dosa,coffee', 100, 'dine-in'),
                                                    (4, '2022-04-01', 'pizza', 200, 'parcel'),
                                                    (5, '2022-04-01', 'pizza,fries', 345, 'parcel'),
                                                    (6, '2022-04-02', 'pizza,sandwich,juice', 567, 'dine-in'),
                                                    (7, '2022-04-02', 'fried rice', 135, 'parcel'),
                                                    (8, '2022-04-02', 'pizza,fries,juice', 550, 'dine-in'),
                                                    (9, '2022-04-02', 'burger,pizza puff,coke', 250, 'dine-in'),
                                                    (10, '2022-04-03', 'paneer tikka,rice', 190, 'parcel'),
                                                    (11, '2022-04-03', 'pizza,fries', 345, 'dine-in')''')

print("All data in food table\n")

# create a cursor object for select query

cursor = connection.execute('select * from restaurant')

# display all data from restaurant table
for row in cursor:
  print(row)
```

### <b>2.executemany():</b>

This method executes an SQL command against all parameter sequences or mappings found in the sequence sql.

<b>Syntax:</b>

connection_name.executemany("sql query",parameters)

* ### Follow the below given steps to insert many records at a time.

1. define a list having tuples of values to be inserted.
2. execute insert command by passing the sql insert query and list of values.
3. execute select command to retreive the data using cursor.

Refer: https://dev.mysql.com/doc/connector-python/en/connector-python-api-mysqlcursor-executemany.html


```python
#define value tuple
values = [(12, '2022-04-03', 'dosa,tea', 140, 'parcel'),
(13, '2022-04-03', 'noodles,sushi', 300, 'dine-in'),
(14, '2022-04-04', 'ice cream,fried rice', 200, 'parcel'),
(15, '2022-04-04', 'chole puri,cake', 250, 'dine-in')]

#execute insert query to insert above tuple
connection.executemany('insert into restaurant values(?,?,?,?,?)',values)

#execute cursor to retrieve data
cursor = connection.execute('select * from restaurant')

# ... and then output all the rows (as tuples):
for row in cursor:
  print(row)
```

### <b>3. fetchone():</b>

This method fetches the next row of a query result set, returning a single sequence, or None when no more data is available.

<b>Syntax:</b>

curosr.fetchone()

Refer: https://dev.mysql.com/doc/connector-python/en/connector-python-api-mysqlcursor-fetchone.html

* ### Display records that have ordered parcel using cursor and fetchone method.



```python
# Execute select statement to display results of parcel records
cursor = connection.cursor()
cursor.execute('''select * from restaurant
                  where Parcel_or_Dine_in = 'parcel' ''')
# Using a while loop
row = cursor.fetchone()
while row is not None:
  print(row)
  row = cursor.fetchone()

# # Using the cursor as iterator
# for row in cursor:
#   print(row)
```

### <b>4.fetchmany:</b>

The method tries to fetch as many rows as indicated by the size parameter.

<b>Syntax:</b>

cursor.fetchmany(size)

Refer: https://dev.mysql.com/doc/connector-python/en/connector-python-api-mysqlcursor-fetchmany.html

* ### <b> Fetch 4 records using fetchmany</b>

```python
# Execute select statement to display results of parcel records
cursor = connection.cursor()
cursor.execute('''select * from restaurant''')

rows = cursor.fetchmany(4)

print(f'''
Fetching Total {len(rows)} rows
Printing each row''')

# Using the cursor as iterator
for r in rows:
  print(f'''\nOrder_ID: {r[0]}
Date: {r[1]}
Items: {r[2]}
Bill_amount: {r[3]}
Parcel/Dine-in: {r[4]}''')
```

### 5.fetchall():</b>

This method fetches all (remaining) rows of a query result, returning a list.

<b>Syntax:</b>

cursor.fetchall()

Refer: https://dev.mysql.com/doc/connector-python/en/connector-python-api-mysqlcursor-fetchall.html

* ### <b> Fetch all records row by row.</b>



```python
cursor = connection.cursor()
cursor.execute('''select * from restaurant''')

# rows = cursor.fetchall() 

head_rows = cursor.fetchmany(size=2)
remaining_rows = cursor.fetchall()


print(f'''
Fetching Total {len(remaining_rows)} rows
Printing each row''')

# Using the cursor as iterator
for r in remaining_rows:
  print(f'''\nOrder_ID: {r[0]}
Date: {r[1]}
Items: {r[2]}
Bill_amount: {r[3]}
Parcel/Dine-in: {r[4]}''')
```
